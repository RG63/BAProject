### Which borough has had the highest total number of crashes reported since 2012? (Ignore missing values)

```python
# Convert to pandas date time
df['CRASH DATE_CRASH TIME'] = pd.to_datetime(df['CRASH DATE_CRASH TIME'])

# Extract the entries that are after 2012 
df2012 = df[df['CRASH DATE_CRASH TIME'].dt.year >= 2012]

# Figure out which borough has the highest total number of crashes
borough_crashes_count = df2012['BOROUGH'].value_counts(dropna=True)

# Get the borough with the maximum number of crashes
max_crash_borough = borough_crashes_count.idxmax()
max_crash_borough
```

### Considering the entire data set since July 2012, which borough has the most crashes for every 100,000 people? 

```python
# Population estimates
populations = {
    'BRONX': 1446788,
    'BROOKLYN': 2648452,
    'MANHATTAN': 1638281,
    'QUEENS': 2330295,
    'STATEN ISLAND': 487155
}

#Filter dataframe for records since July 2012
df['CRASH DATE_CRASH TIME'] = pd.to_datetime(df['CRASH DATE_CRASH TIME'])
df_since_july_2012 = df[df['CRASH DATE_CRASH TIME'] >= '2012-07-01']

#Count the Number of Crashes in Each Borough (ignoring NaN values)
borough_crash_counts = df_since_july_2012['BOROUGH'].value_counts(dropna=True)

#Calculate the Rate of Crashes per 100,000 People for Each Borough
crash_rates = {}
for borough, crash_count in borough_crash_counts.items():
    if borough in populations:  # Check if the borough is in our population dictionary
        crash_rate = (crash_count / populations[borough]) * 100000
        crash_rates[borough] = crash_rate

#Find the Borough with the Maximum Rate
max_rate_borough = max(crash_rates, key=crash_rates.get)
print(f"{max_rate_borough} has the most crashes for every 100,000 people with a rate of {crash_rates[max_rate_borough]:.2f}.")
```

### Analyzing the leading cause of crashes

```python
#Filter the DataFrame to Exclude 'Unspecified' Values
filtered_df = df[df['CONTRIBUTING FACTOR VEHICLE 1'] != 'Unspecified']

#Find the Leading Cause of Crashes
leading_cause = filtered_df['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().idxmax()

#Calculate the Proportion of Accidents Attributable to this Cause
total_crashes = len(filtered_df)
leading_cause_crashes = filtered_df['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().max()
proportion = (leading_cause_crashes / total_crashes) * 100  # to convert to percentage

#Display the Results
print(f"The leading cause of crashes is '{leading_cause}'")
print(f"The proportion of accidents attributable to this cause is {proportion:.2f}%")
```

### Top 3 causes of crashes 

```python
#Filter the DataFrame to Exclude 'Unspecified' Values
filtered_df = df[df['CONTRIBUTING FACTOR VEHICLE 1'] != 'Unspecified']

#Find the Top 3 Causes of Crashes
top_3_causes = filtered_df['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().head(3)

#Calculate the Proportion of Accidents Attributable to the Top 3 Causes
total_crashes = len(filtered_df)
top_3_crashes = top_3_causes.sum()
proportion_top_3 = (top_3_crashes / total_crashes) * 100  # to convert to percentage

print(f"The top 3 causes of crashes are {', '.join(top_3_causes.index)}")
print(f"The proportion of accidents attributable to these top 3 causes is {proportion_top_3:.2f}%")
```

### Analyzing fatalities

```python
#Filter the DataFrame for Rows Where NUMBER OF PERSONS KILLED > 0
fatal_accidents_df = df[df['NUMBER OF PERSONS KILLED'] > 0]

#Count the Number of Rows in the Filtered DataFrame
count_fatal_accidents = len(fatal_accidents_df)

print(f"The total count of accidents that involved at least one fatality is {count_fatal_accidents}.")
```

### Likelihood of fatal accidents
#### On average, out of every 1000 accidents, how many have resulted in at least one person dead?

```python
#Count the Total Number of Accidents
total_accidents = len(df)

#Count the Number of Accidents with at least one Fatality
fatal_accidents_df = df[df['NUMBER OF PERSONS KILLED'] > 0]
count_fatal_accidents = len(fatal_accidents_df)
 
#Calculate the Ratio and Convert to per 1000 Accidents
ratio = (count_fatal_accidents / total_accidents) * 1000

#Display the Result
print(f"On average, for every 1000 accidents, {ratio:.3f} have resulted in at least one person dead.")
```

### Missing data
#### What proportion of accidents in the data do not have a Borough code?

```python
#Count the Total Number of Accidents
total_accidents = len(df)

#Count the Number of Accidents with Missing Borough Code
missing_borough_count = df['BOROUGH'].isna().sum()

#Calculate the Proportion
proportion_missing_borough = (missing_borough_count / total_accidents) * 100

print(f"The proportion of accidents that do not have a Borough code is {proportion_missing_borough:.2f}%.")
```

### Which combinations of vehicles have the most number of accidents

```python
#Group by 'VEHICLE TYPE CODE 1' and 'VEHICLE TYPE CODE 2'
vehicle_combination_counts = df.groupby(['VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2']).size().reset_index(name='count')

#Find the Combination with the Most Accidents
max_combination = vehicle_combination_counts.loc[vehicle_combination_counts['count'].idxmax()]

print(f"The combination of vehicles with the most number of accidents is {max_combination['VEHICLE TYPE CODE 1']} and {max_combination['VEHICLE TYPE CODE 2']} with {max_combination['count']} accidents.")
```
